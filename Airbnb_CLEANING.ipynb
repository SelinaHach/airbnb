```python
import pandas as pd

df_boston = pd.read_csv("/content/Boston listings.csv")
df_dc = pd.read_csv("/content/DC listings.csv")
df_la = pd.read_csv("/content/LA listings.csv")
df_ny = pd.read_csv("/content/New York listings.csv")
df_sf = pd.read_csv("/content/San Fran listings.csv")
df_seattle = pd.read_csv("/content/Seattle listings.csv")
```

```python
dfs = []
dfs.append(df_boston.assign(city='Boston'))
dfs.append(df_dc.assign(city='Washington DC'))
dfs.append(df_la.assign(city='Los Angeles'))
dfs.append(df_ny.assign(city='New York'))
dfs.append(df_sf.assign(city='San Francisco'))
dfs.append(df_seattle.assign(city='Seattle'))

df_combined = pd.concat(dfs, ignore_index=True)
display(df_combined.head())
display(df_combined.shape)
```

```python
# Select relevant columns
relevant_columns = [
    'host_name', 'listing_url', 'price', 'last_scraped', 'host_id', 'host_since', 'host_response_time',
    'host_response_rate', 'host_is_superhost', 'host_listings_count',
    'host_identity_verified', 'neighbourhood_group_cleansed', 'neighbourhood_cleansed', 'room_type',
    'accommodates', 'bathrooms', 'bedrooms', 'minimum_nights', 'number_of_reviews',
    'review_scores_rating', 'estimated_occupancy_l365d', 'last_review', 'instant_bookable', 'city' # Include city for analysis
]
df_cleaned = df_combined[relevant_columns].copy()

# Inspect data types (initial inspection after selection)
print("Initial data types after column selection:")
print(df_cleaned.dtypes)

# Clean and convert 'price'
df_cleaned['price'] = df_cleaned['price'].astype(str).str.replace(r'[$,]', '', regex=True)
# Convert to numeric, coercing errors to NaN to handle non-numeric values
df_cleaned['price'] = pd.to_numeric(df_cleaned['price'], errors='coerce')


# Convert 'host_response_rate'
df_cleaned['host_response_rate'] = df_cleaned['host_response_rate'].astype(str).str.replace('%', '', regex=False)
# Convert to numeric, coercing errors to NaN
df_cleaned['host_response_rate'] = pd.to_numeric(df_cleaned['host_response_rate'], errors='coerce') / 100


# Clean and convert 'bathrooms'
# Attempt to convert to numeric directly, coercing errors to NaN
df_cleaned['bathrooms'] = pd.to_numeric(df_cleaned['bathrooms'], errors='coerce')

# Convert date columns to datetime
df_cleaned['last_scraped'] = pd.to_datetime(df_cleaned['last_scraped'])
df_cleaned['host_since'] = pd.to_datetime(df_cleaned['host_since'])

# Standardize boolean columns
df_cleaned['host_is_superhost'] = df_cleaned['host_is_superhost'].map({'t': True, 'f': False}).astype(bool)
df_cleaned['host_identity_verified'] = df_cleaned['host_identity_verified'].map({'t': True, 'f': False}).astype(bool)
df_cleaned['instant_bookable'] = df_cleaned['instant_bookable'].map({'t': True, 'f': False}).astype(bool)

# Check missing values after cleaning and type conversion
print("\nMissing values after cleaning and type conversion:")
print(df_cleaned[relevant_columns].isnull().sum())

display(df_cleaned.head())
display(df_cleaned.dtypes)
```

```python
# Calculate host experience in years
df_cleaned['host_duration_years'] = (df_cleaned['last_scraped'] - df_cleaned['host_since']).dt.days / 365.25

# Handle missing values in the new feature
# Fill NaN values with -1 as an indicator of missing host_since data
df_cleaned['host_duration_years'].fillna(-1, inplace=True)

# Display the first few rows with the new feature
display(df_cleaned[['host_since', 'last_scraped', 'host_duration_years']].head())
```

```python
# Check for NAs per city and per column
na_by_city_and_column = df_cleaned.isna().groupby(df_cleaned['city']).sum()

# Adjust display options to show all columns
pd.set_option('display.max_columns', None)
pd.set_option('display.expand_frame_repr', False)

# Display the results for NAs
print("Missing values per variables per city:")
display(na_by_city_and_column)

# Get the total number of entries per city
total_entries_per_city = df_cleaned['city'].value_counts()

# Calculate the total number of NAs per city across all columns
total_NAs_per_city = na_by_city_and_column.sum(axis=1)

# Display the total number of entries per city
print("\nTotal number of entries per city:")
display(total_entries_per_city)

# Display the total number of NAs per city across all columns
print("\nTotal number of NAs per city:")
display(total_NAs_per_city)


# Reset display options to default
pd.reset_option('display.max_columns')
pd.reset_option('display.expand_frame_repr')
```

```python
# Fill NA values in 'neighbourhood_group_cleansed' with values from 'neighbourhood_cleansed'
df_cleaned['neighbourhood_group_cleansed'] = df_cleaned['neighbourhood_group_cleansed'].fillna(df_cleaned['neighbourhood_cleansed'])

# Recalculate NA values per city and per column after filling
na_by_city_and_column_updated = df_cleaned.isna().groupby(df_cleaned['city']).sum()

# Display the updated results for NAs
print("Missing values per variables per city (after filling neighbourhood_group_cleansed):")
display(na_by_city_and_column_updated)
```

```python
# Drop the original 'neighbourhood_cleansed' column
df_cleaned = df_cleaned.drop('neighbourhood_cleansed', axis=1)
```

```python
# Identify columns with missing values across variables
cols_with_na = df_cleaned.columns[df_cleaned.isna().any()].tolist()

# Check for rows where all columns that have NAs are missing
rows_missing_all_cols_with_na = df_cleaned[df_cleaned[cols_with_na].isna().all(axis=1)]

print(f"Checking for rows where all of the following columns are missing: {cols_with_na}")
print("\nRows with missing values in ALL columns that contain any NAs:")
display(rows_missing_all_cols_with_na)

# You can also check how many such rows exist
print("\nNumber of rows with missing values in ALL columns that contain any NAs:")
print(rows_missing_all_cols_with_na.shape[0])
```

```python
# Identify rows where all columns that have NAs are missing in the current df_cleaned
rows_to_drop = df_cleaned[df_cleaned[cols_with_na].isna().all(axis=1)].index

print(f"Identified {len(rows_to_drop)} rows to drop where all columns with NAs were missing.")

# Drop the identified rows
df_cleaned = df_cleaned.drop(rows_to_drop)

print(f"Dropped {len(rows_to_drop)} rows.")
print("\nShape of df_cleaned after dropping rows:")
display(df_cleaned.shape)
```

```python
# Drop New York data
df_cleaned = df_cleaned[df_cleaned['city'] != 'New York'].copy()

# Recalculate NA values per city and per column after filling cleaning
na_by_city_and_column_updated = df_cleaned.isna().groupby(df_cleaned['city']).sum()


# Adjust display options to show all columns
pd.set_option('display.max_columns', None)
pd.set_option('display.expand_frame_repr', False)

# Display the updated results for NAs
print("Missing values per variables per city after cleaning:")
display(na_by_city_and_column_updated)

# Reset display options to default
pd.reset_option('display.max_columns')
pd.reset_option('display.expand_frame_repr')

# Check for NAs per city and per column
na_by_city_and_column_updated = df_cleaned.isna().groupby(df_cleaned['city']).sum()
# Calculate the total number of NAs per city across all columns
total_NAs_per_city_updated = na_by_city_and_column_updated.sum(axis=1)


# Display the total number of NAs per city across all columns
print("\nTotal number of NAs per city:")
display(total_NAs_per_city_updated)



# Get the total number of entries per city
total_entries_per_city_updated = df_cleaned['city'].value_counts()

# Display the total number of entries per city
print("\nTotal number of entries per city:")
display(total_entries_per_city_updated)
```

```python
# --- Per-city quartile and bounds summary ---
iqr_summary = df_cleaned.groupby('city')['price'].describe(percentiles=[0.25, 0.75]).reset_index()

# Compute IQR and bounds directly
iqr_summary['IQR'] = iqr_summary['75%'] - iqr_summary['25%']
iqr_summary['Lower_Bound'] = iqr_summary['25%'] - 1.5 * iqr_summary['IQR']
iqr_summary['Upper_Bound'] = iqr_summary['75%'] + 1.5 * iqr_summary['IQR']

# --- Count listings above upper bound per city ---
merged = df_cleaned.merge(
    iqr_summary[['city', 'Upper_Bound']],
    on='city',
    how='left'
)

above_upper = (
    merged[merged['price'] > merged['Upper_Bound']]
    .groupby('city')
    .size()
    .reset_index(name='# Listings above upper bound')
)

# Merge count into summary table
summary_table = (
    iqr_summary.merge(above_upper, on='city', how='left')
    [['city', '25%', '75%', 'IQR', 'Lower_Bound', 'Upper_Bound', '# Listings above upper bound']]
    .rename(columns={'25%': 'Q1', '75%': 'Q3'})
)

# Fill missing values for cities with no outliers
summary_table['# Listings above upper bound'] = summary_table['# Listings above upper bound'].fillna(0).astype(int)

# --- Display neatly (center-aligned) ---
display(
    summary_table
        .style
        .format({
            'Q1': '{:.1f}',
            'Q3': '{:.1f}',
            'IQR': '{:.1f}',
            'Lower_Bound': '{:.1f}',
            'Upper_Bound': '{:.1f}',
            '# Listings above upper bound': '{:,.0f}'
        })
        .set_caption("Price Outlier Range by City")
        .set_properties(**{'text-align': 'center'})  # centers all text
        .set_table_styles([{
            'selector': 'th',
            'props': [('text-align', 'center')]  # centers column headers
        }])
)
```

```python
# 1. Read manual outliers
outliers = pd.read_excel("BSAN460_Outliers_Combined.xlsx")

# 2. Make sure the join key has the same type in both
df_cleaned['listing_url'] = df_cleaned['listing_url'].astype(str).str.strip()
outliers['listing_url'] = outliers['listing_url'].astype(str).str.strip()

# 3. Anti-join on listing_url only
merged = df_cleaned.merge(
    outliers[['listing_url']],   # join key
    on='listing_url',
    how='left',
    indicator=True
)

df_cleaned_filtered = merged[merged['_merge'] == 'left_only'].drop(columns=['_merge'])

# 4. How many were actually removed
n_removed_within_upper_bound = df_cleaned.shape[0] - df_cleaned_filtered.shape[0]
print(f"# Rows removed based on manual outliers checked (by listing_url): {n_removed_within_upper_bound}")

# 5. Per-city breakdown for rows that WERE removed
removed_per_city = (
    merged[merged['_merge'] == 'both']   # rows that matched the outlier file
    .groupby('city')
    .size()
    .reset_index(name='n_removed_within_upper_bound')
    .sort_values('n_removed_within_upper_bound', ascending=False)
)

print("\nRemoved listings within upper-bound:")
display(
    removed_per_city.style
        .set_properties(**{'text-align': 'center'})
        .set_table_styles([{'selector': 'th','props':[('text-align','center')]}])
        .format({'n_removed_within_upper_bound': '{:,.0f}'})
)

# 6. Overwrite df_cleaned
df_cleaned = df_cleaned_filtered
```
